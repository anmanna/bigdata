```python
import pandas as pd
import matplotlib.pyplot as plt

def time_plt(grape, title):
  grape['time'] = pd.to_datetime(grape['time'])
  fig, axs = plt.subplots(figsize=(12, 6))

  grape.groupby(grape['time'].dt.year)['prediction'].mean().plot(
    kind='line', color='black', ax=axs, title=''
  )
  plt.title(title)
  plt.ylabel('Sentiment')
  plt.xlabel('Year')

chardonnay = pd.read_csv('/content/drive/Shareddrives/Big data/nb/barbera_predict.csv', lineterminator='\n')#, low_memory=False
chardonnay['id'] = chardonnay['id'].astype(str)
chardonnay['time'] = chardonnay['time'].astype(str)
chardonnay = chardonnay.dropna()
time_plt(chardonnay, 'chardonnay')
```
