# Functions
Read files, merge and plot choropleth world maps

```python
def read_file(file_location):
  data_frame = pd.read_csv(file_location, lineterminator='\n')

  return data_frame
```

```python
def join(csv1, csv2):
  df = csv1.merge(csv2, on='id')

  return df
```

```python
def get_country(df):

  df['location'] = df['location'].str.rsplit(',')
  
  df['country'] = df['location']

  for i in range (len(df['location'])):
    df['location'][i].reverse()
    df['country'][i] = df['location'][i][0]

  return df
```

https://levelup.gitconnected.com/plotting-choropleth-maps-in-python-b74c53b8d0a6

```python
def plot_chm(df, title):
  chm = go.Figure(
    data = {
        'type':'choropleth',
        'locations':df['country'],
        'locationmode':'country names',
        'colorscale':'Jet',
        'z':df['prediction'],
        'colorbar':{'title':'Sentiment of each country'},
        'marker': {
            'line': {
                'color':'rgb(0,0,0)',
                'width':1
            }
        }
    },    

    layout = {    
      'title': title,      
      'geo':{
          'scope':'world', 
      }
    })

  return chm

```

Functions and plots

```python
def time_plt(grape, title):
  grape['time'] = pd.to_datetime(grape['time'])
  fig, axs = plt.subplots(figsize=(12, 6))

  grape.groupby(grape['time'].dt.year)['prediction'].mean().plot(
    kind='line', color='black', ax=axs, title=''
  )
  plt.title(title)
  plt.ylabel('Sentiment')
  plt.xlabel('Year')
```
